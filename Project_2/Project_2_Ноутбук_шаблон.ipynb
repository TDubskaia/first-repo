7. Для компании «Яндекс» выведите список регионов-миллионников, в которых представлены вакансии компании, вместе с количеством вакансий в этих регионах. Также добавьте строку Total с общим количеством вакансий компании. Результат отсортируйте по возрастанию количества.

Список городов-милионников надо взять [отсюда](https://ru.wikipedia.org/wiki/%D0%93%D0%BE%D1%80%D0%BE%D0%B4%D0%B0-%D0%BC%D0%B8%D0%BB%D0%BB%D0%B8%D0%BE%D0%BD%D0%B5%D1%80%D1%8B_%D0%A0%D0%BE%D1%81%D1%81%D0%B8%D0%B8). 

Если возникнут трудности с этим задание посмотрите материалы модуля  PYTHON-17. Как получать данные из веб-источников и API. 

```python
# код для получения списка городов-милионников
import requests
from bs4 import BeautifulSoup

#получаем данные со страницы википедии
url = 'https://ru.wikipedia.org/wiki/%D0%93%D0%BE%D1%80%D0%BE%D0%B4%D0%B0-%D0%BC%D0%B8%D0%BB%D0%BB%D0%B8%D0%BE%D0%BD%D0%B5%D1%80%D1%8B_%D0%A0%D0%BE%D1%81%D1%81%D0%B8%D0%B8'
response = requests.get(url)
page = BeautifulSoup(response.text,'html.parser')

#находим нужную таблицу в разметке страницы
cities_here = str(page.find('table', class_="standard")) 
cities_here = cities_here.split('title="')[3:] #получаем список говородов с другой информацией, от который далее избавимся

cities = []
for i in cities_here:
    cities.append(i.split('">')[0]) #получаем только первый элемент, после разделения строки, это и будет нужный нам город
cities = tuple(cities) #преобразуем список в кортеж   
```

```python
# текст запроса
query_5_7 = f'''
            with temp as( 
            --вложенный запрос, с которым далее работаем, как с таблицей temp
                select 
                    a.name,
                    count(v.id) cnt_v
                from vacancies v
                    join areas a on v.area_id = a.id              --соединение с таблицей регионов, чтобы узнать название региона
                    join employers e on v.employer_id = e.id      --соединение с таблицей работодателей, чтобы узнать название работодателя
                where e.name like 'Яндекс' and a.name in {cities} --работаем только с организацией Яндекс и регионами-городами-милионниками
                group by a.name
            )
            select * from temp --выбираем все из вложенного запроса
            
            UNION 
            
            select
                'Total',
                sum(cnt_v) --считаем общее количество вакансий без группировки по региону
            from temp
            order by 2 desc --сортируем по количеству вакансий
'''
```

```python
# результат запроса
df = pd.read_sql_query(query_5_7, connection)
df
```

При решении задачи примите во внимание следующее:
1. Рассматриваем только вакансии, у которых заполнено хотя бы одно из двух полей с зарплатой.
2. Если заполнены оба поля с зарплатой, то считаем зарплату по каждой вакансии как сумму двух полей, делённую на 2. Если заполнено только одно из полей, то его и считаем зарплатой по вакансии.
3. Если в расчётах участвует null, в результате он тоже даст null (посмотрите, что возвращает запрос select 1 + null). Чтобы избежать этой ситуацию, мы воспользуемся функцией [coalesce](https://postgrespro.ru/docs/postgresql/9.5/functions-conditional#functions-coalesce-nvl-ifnull), которая заменит null на значение, которое мы передадим. Например, посмотрите, что возвращает запрос `select 1 + coalesce(null, 0)`

Выясните, на какую зарплату в среднем может рассчитывать дата-сайентист с опытом работы от 3 до 6 лет. Результат округлите до целого числа. 

```python
# текст запроса
query_6_6 = f'''
            select 
                experience,  
                round(avg(coalesce((salary_from + salary_to)/2, salary_from, salary_to)), 0) salary_avg
            from vacancies
            --ниже добавляем условия для отбора вакансий для дата-сайентистов
            where (lower(name) like '%data scientist%' 
                or lower(name) like '%data science%'
                or lower(name) like '%исследователь данных%'
                or (name like '%ML%' and not name like '%HTML%')
                or lower(name) like '%machine learning%'
                or lower(name) like '%машинн%обучен%')
                and (not salary_from is null or not salary_to is null) --выбираем те записи, где указана заработная плата
            group by experience --группируем по опыту   
'''
```
